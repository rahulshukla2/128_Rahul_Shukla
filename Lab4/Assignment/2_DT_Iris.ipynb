```python
#Import scikit-learn dataset library
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import numpy as np

#Load dataset
iris = datasets.load_iris()
```

```python
# print the names of the 13 features

print(iris.feature_names)


# print the label type of wine(class_0, class_1, class_2)
print(iris.target_names)

# print data(feature)shape
X=iris.data
y=iris.target
print(X.shape)
```

```python
#import the necessary module
from sklearn.model_selection import train_test_split

#split data set into train and test sets
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.4, random_state=68)
```

```python
#Create a Decision Tree Classifier (using Gini)
clf_gini=DecisionTreeClassifier(criterion="gini")

#Train the model using the training sets
clf_gini.fit(X_train,y_train)
```

```python
# Predict the classes of test data
test_pred=clf_gini.predict(X_test)
print(test_pred.dtype)
from sklearn import metrics
```

```python
# Model Accuracy, how often is the classifier correct?
from sklearn.metrics import accuracy_score
print('\nAccuracy: {0:.4f}'.format(accuracy_score(y_test, test_pred)))
```

```python
from sklearn.tree import export_graphviz
export_graphviz(clf_gini,out_file='iris_tree.dot',feature_names=list(iris.feature_names),
                class_names=list(iris.target_names), filled=True)

# Convert to png
from subprocess import call
call(['dot', '-Tpng', 'iris_tree.dot', '-o', 'iris_tree.png', '-Gdpi=600'])
```

```python
# Display in python
import matplotlib.pyplot as plt
plt.figure(figsize = (14, 18))
plt.imshow(plt.imread('iris_tree.png'))
plt.axis('off')
plt.show()
```
